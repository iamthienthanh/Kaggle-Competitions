```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
train
```

```python
test = pd.read_csv('test.csv')
test
```

```python
women_survived = train.loc[train.Sex == "female"]["Survived"]
f"% of women survived: {sum(women_survived)/len(women_survived)}"
```

```python
men_survived = train.loc[train.Sex == "male"]["Survived"]
f"% of men survived: {sum(men_survived)/len(men_survived)}"
```

```python
# import the `RandomForestClassifier`
from sklearn.ensemble import RandomForestClassifier

features = ["Pclass", "Sex", "SibSp", "Parch"]

y_train = train["Survived"]
x_train = pd.get_dummies(train[features])
x_test = pd.get_dummies(test[features])
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

# evaluate the model
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x_train, y_train, cv=5)
scores.mean()
```

```python
# Create submission file
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': y_pred})
output.to_csv('submission.csv', index=False)
```
